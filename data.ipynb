```python
import os
import zipfile
import pandas as pd
```

```python
main_folder = 'data'  
subfolders = [  
    'raw_match_parquet',  
    'raw_odds_parquet',  
    'raw_point_by_point_parquet',  
    'raw_statistics_parquet',  
    'raw_tennis_power_parquet',  
    'raw_votes_parquet'  
]  
```

```python
dataframes = []  
for date in range(20240501, 20240532):  
    date_folder = os.path.join(main_folder, str(date))   
    if os.path.exists(date_folder):  
        for subfolder in subfolders:  
            subfolder_path = os.path.join(date_folder, subfolder)  
            if os.path.exists(subfolder_path):  
                for file in os.listdir(subfolder_path):  
                    if file.endswith('.parquet'):  
                        file_path = os.path.join(subfolder_path, file)  
                        try:  
                            df = pd.read_parquet(file_path)   
                            if all(col in df.columns for col in [
'match_id',
'first_to_serve',
'home_team_seed',   
'away_team_seed',
'custom_id',
'winner_code',   
'default_period_count',
'start_datetime',
'match_slug',   
'final_result_only']):  
                                selected_columns = df[[
'match_id',
'first_to_serve',
'home_team_seed',   
'away_team_seed',
'custom_id',
'winner_code',   
'default_period_count',
'start_datetime',
'match_slug',   
'final_result_only']]  
                                dataframes.append(selected_columns)  
                        except Exception as e:  
                            print(f"خطا در خواندن فایل {file_path}: {e}")  
if dataframes:  
    MatchEventInfo = pd.concat(dataframes, ignore_index=True)  
    print("DataFrame MatchEventInfo ساخته شد.")  
else:  
    MatchEventInfo = pd.DataFrame(columns=[
'match_id',
'first_to_serve',
'home_team_seed',   
'away_team_seed',
'custom_id',
'winner_code',   
'default_period_count',
'start_datetime',
'match_slug',   
'final_result_only'])  
    print("هیچ داده‌ای برای ساخت DataFrame وجود ندارد.")  
print(MatchEventInfo.head())
```
